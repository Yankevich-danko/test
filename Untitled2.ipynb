```python
1.5 Кластеризация
```

```python
#Помещу их в переменную X
X = df_merhe[["AGE", "GENDER", "EDUCATION_LEVEL"]]
```

```python
#обьявляю MinMaxScaler
scaler = MinMaxScaler()
#преобразую данные
X = scaler.fit_transform(X)
```

```python
#Обьявляю метод главных компонент
pca = PCA(n_components=2)
#Применяю его на данных
X = pca.fit_transform(X)
```

```python
#Функция для визуализации распределения 
def viz(prediction):
    #Размер фигуры
    plt.figure(figsize=(12, 12))
    plt.subplot(224)
    #Выводить изображение буду при помощи scatter
    plt.scatter(X[:, 0], X[:, 1], c=prediction)
    plt.title("Unevenly Sized Blobs")
    #Вывод изображения
    plt.show()
```

```python
#Настройка параметров
kmeans = KMeans(n_clusters=4, random_state=0).fit(X)
#Предсказание
kmpreds = kmeans.predict(X)
#Заношу кластеризированные метки в набор данных
df_merhe["KMCLUSTS"] = kmpreds

#Применяю функцию описаную выше
viz(df_merhe["KMCLUSTS"])
```

```python
#Настройка параметров
kbmeans = MiniBatchKMeans(n_clusters=4, random_state=0, batch_size=4096)
#Заношу кластеризированные метки в набор данных
df_merhe["KBCLUSTS"] = kbmeans.fit_predict(X)

viz(df_merhe["KBCLUSTS"])
```

```python
#Настройка параметров
gm = GaussianMixture(n_components=2, random_state=0).fit_predict(X)
#Заношу кластеризированные метки в набор данных
df_merhe["GMCLUSTS"] = gm

viz(df_merhe["GMCLUSTS"])
```

```python
#Использую метрику calinski_harabasz_score
print(sklearn.metrics.calinski_harabasz_score(X, df_merhe["KMCLUSTS"]))
#Использую метрику davies_bouldin_score
print(sklearn.metrics.davies_bouldin_score(X, df_merhe["KMCLUSTS"]))
```

```python
#Использую метрику calinski_harabasz_score
print(sklearn.metrics.calinski_harabasz_score(X, df_merhe["KBCLUSTS"]))
#Использую метрику davies_bouldin_score
print(sklearn.metrics.davies_bouldin_score(X, df_merhe["KBCLUSTS"]))
```

```python
#Использую метрику calinski_harabasz_score
print(sklearn.metrics.calinski_harabasz_score(X, df_merhe["GMCLUSTS"]))
#Использую метрику davies_bouldin_score
print(sklearn.metrics.davies_bouldin_score(X, df_merhe["GMCLUSTS"]))
```

```python
n_estimators=[64,100,128,200]
max_features= [2,3,4]
bootstrap = [True,False]
oob_score = [True,False]
```

```python
rfc = RandomForestClassifier()
grid = GridSearchCV(rfc,param_grid)
```

```python
grid.fit(X_train,y_train)
```

```python
predictions = grid.predict(X_test)
```

```python
errors = []
misclassifications = []

for n in range(1,64):
    rfc = RandomForestClassifier( n_estimators=n,bootstrap=True,max_features= 2)
    rfc.fit(X_train,y_train)
    preds = rfc.predict(X_test)
    err = 1 - accuracy_score(preds,y_test)
    n_missed = np.sum(preds != y_test) # Более подробно об этой строке кода см. в видео-лекции!!
    errors.append(err)
    misclassifications.append(n_missed)
```
